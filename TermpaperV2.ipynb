```python
def plot_corr(df, size=15):
    corr = df.corr()
    fig, ax = plt.subplots(figsize=(size, size))
    ax.matshow(corr)
    plt.xticks(range(len(corr.columns)), corr.columns)
    plt.yticks(range(len(corr.columns)), corr.columns)
    
plot_corr(diadata) 
```

Let's look now at the distribution of the Outcome to see if the distribution of Positive/Negative (True/False) is adequate for our future analysis.

```python
fig, axes = plt.subplots(1, 1, figsize=(10, 10))
explode = (0, 0.1)
axes.pie(diadata['Outcome'].value_counts(), labels=diadata['Outcome'].value_counts().index.tolist(), autopct='%1.1f%%', 
        shadow=True, startangle=90, explode = explode )
axes.set_title('Outcome')
axes.legend()
```

The distrubution shows that 34.9% of our patients have diabetes and 65.1% have not. This is a good distribution. 

Let's now slit our dataset in "training data" and "testing data". We will use a 75/25 split.

```python
from sklearn.model_selection import train_test_split

features_cols = ['Pregnancies', 'Glucose', 'BloodPressure', 'SkinThickness', 'Insulin', 'BMI', 
                 'DiabetesPedigreeFunction', 'Age']
predicted_class = ['Outcome']

X = diadata[features_cols].values    
Y = diadata[predicted_class]. values 
split_test_size = 0.25

x_train, x_test, y_train, y_test = train_test_split(X, Y, test_size=split_test_size, random_state=1)

print("{0:0.2f}% in training set".format((len(x_train)/len(diadata.index)) * 100))
print("{0:0.2f}% in test set".format((len(x_test)/len(diadata.index)) * 100))
```

Looks like we have a "prefect" split of 75/25 are we planned.



Logistic Regression models are good models for categorical (binary) dependant variable which is our case. 

LOGISTIC REGRESSION

```python
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import accuracy_score
from sklearn.metrics import confusion_matrix

LR = LogisticRegression()
LR.fit(x_train, y_train.ravel())
LR_predictions = LR.predict(x_test)

print(accuracy_score(y_test, LR_predictions))
```

Let's look not a the confusion matrix

```python
print(confusion_matrix(y_test, LR_predictions, labels=[1, 0]))
```

Also we can look at the Recall (also known as sensitivity) which is the fraction of positives events that is predicted correctly by the model

```python
from sklearn.metrics import recall_score
print (recall_score(y_test, LR_predictions))
```

And finally the Precision whici is the fraction of predicted positives events that are actually positive

```python
from sklearn.metrics import precision_score
print(precision_score(y_test, LR_predictions))
```
